```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

# Set up Selenium WebDriver with Chrome
driver = webdriver.Chrome()  # Uses the default Chrome driver location

# Define the website and URL
website = "openinapp.com"
max_results = 1000  # Maximum number of results
url = f"https://www.google.com/search?q=site:youtube.com%20{website}"

# Navigate to the Google search page
driver.get(url)

# Initialize variables
results_count = 0  # Counter for the number of results

# Function to click the "Next" button
def click_next_button():
    next_button = driver.find_element(By.XPATH, "//a[@id='pnnext']")
    if next_button:
        next_button.click()

# Parse and scrape the search results
while results_count < max_results:
    # Wait for the search results to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.g")))

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Find all search result elements
    search_results = soup.select("div.g")

    # Extract and print the YouTube channel names
    for result in search_results:
        channel_name_element = result.select_one("div.P7xzyf")
        if channel_name_element:
            channel_name = channel_name_element.get_text()
            print("YouTube Channel Name:", channel_name)
            results_count += 1

            if results_count >= max_results:
                break

    # Check if the maximum number of results is reached
    if results_count >= max_results:
        break

    # Click the "Next" button
    click_next_button()

# Quit the WebDriver
driver.quit()
```

```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import csv
import json
import re

# Set up Selenium WebDriver with Chrome
driver = webdriver.Chrome()

# Define the search query and URL
search_query = "site:youtube.com openinapp.com"
num_results = 200  # Maximum number of results
url = f"https://www.google.com/search?q={search_query}&num=100"

# Navigate to the Google search page
driver.get(url)

# Initialize variables
channel_names = []
channel_links = []
page_count = 1  # Counter for the number of pages

# Function to click the "More results" button
def click_more_results():
    try:
        more_results_button = driver.find_element(By.XPATH, "//div[@id='smb']")
        if more_results_button:
            driver.execute_script("arguments[0].click();", more_results_button)
    except NoSuchElementException:
        pass

# Parse the HTML content and extract channel names and links
while len(channel_names) < num_results:
    # Wait for the search results to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.g")))

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Find all search result elements
    search_results = soup.select("div.g")

    # Extract the YouTube channel names and links from the search results
    for result in search_results:
        channel_name_element = result.select_one("div.P7xzyf")
        if channel_name_element:
            channel_text = channel_name_element.get_text()
            channel_name = channel_text.split("·")[1].strip()
            modified_channel_name = re.sub(r'[^a-zA-Z0-9\s]', '', channel_name)  # Keep alphanumeric characters and spaces
            modified_channel_name = modified_channel_name.replace(" ", "")  # Remove spaces
            modified_channel_name = modified_channel_name.lower()  # Convert to lowercase
            channel_link = f"https://youtube.com/{modified_channel_name}"
            channel_names.append(channel_name)
            channel_links.append(channel_link)

            if len(channel_names) >= num_results:
                break

    # Check if the maximum number of results is reached
    if len(channel_names) >= num_results:
        break

    # Check if there is a "More results" button and click it
    click_more_results()

# Quit the WebDriver
driver.quit()

# Save the channel names and links in JSON format
channel_data = [{'name': name, 'link': link} for name, link in zip(channel_names, channel_links)]
with open("channel_data.json", "w") as json_file:
    json.dump(channel_data, json_file)

# Save the channel names and links in CSV format
with open("channel_data.csv", "w", newline="", encoding="utf-8") as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["Channel Name", "Channel Link"])
    writer.writerows([[name, link] for name, link in zip(channel_names, channel_links)])
```
